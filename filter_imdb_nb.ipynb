```python
import pandas as pd
# this input file is tab-delimited instead of comma-delimited
# we can specify the delimiter by using the sep keyword argument
title_data = pd.read_csv("title.basics.tsv", sep='\t')
title_data.columns
```

```python
for col in title_data.columns:
    if (col != "endYear" ):
        title_data= title_data[title_data[col] != "\\N"]
```

```python
title_data
```

```python
title_data.dropna()
```

```python
titleTypes_we_want = ["movie", "tvSeries", "tvEpisode"]
```

```python
title_data = title_data.drop(columns=["originalTitle"], errors='ignore')
title_data = title_data[title_data["titleType"].isin(titleTypes_we_want)]
title_data = title_data.rename(columns={"genres": "genre"})
title_data
```

```python
f = lambda x: x["genre"].split(",")[0]
title_data['genre'] = title_data.apply(f, axis=1)
```

```python
genres_we_want = ["Action", "Adventure", "Comedy", "Drama", "Fantasy", "Mystery", "Thriller", "Crime", "Horror", "Historical Fiction", "Political", "Romance", "Science Fiction", "Superhero"]
title_data = title_data[title_data["genre"].isin(genres_we_want)]
title_data
```

```python
title_data["genre"].unique()
```

```python
f = open("titles.csv", 'w')
title_data.to_csv(f)
f.close()
```

```python
tconsts = title_data["tconst"].unique()
tconsts
```

```python
principals_data = pd.read_csv("title.principals.tsv", "\t")
principals_data
```

```python
professions_we_want = ["actor", "actress", "director", "producer"]
```

```python
principals_data = principals_data[principals_data["tconst"].isin(tconsts)]
principals_data = principals_data.drop(columns={"ordering","job","characters"}, errors='ignore')
principals_data = principals_data[principals_data["category"].isin(professions_we_want)]
principals_data
```

```python
f = open("principals.csv", 'w')
principals_data.to_csv(f)
f.close()
```

```python
names_data = pd.read_csv("name.basics.tsv", "\t")
names_data
```

```python
names_data = names_data[names_data["nconst"].isin(nconsts)]
names_data = names_data.drop(columns={"knownForTitles"})
names_data
```

```python
names_data = names_data[names_data["primaryProfession"].notnull()]
f = lambda x: x["primaryProfession"].split(",")[0]
names_data["primaryProfession"] = names_data.apply(f, axis=1)
names_data
```

```python
professions_we_want = ["actor", "actress", "director", "producer", "writer", "soundtrack", "music_department"]
names_data = names_data[names_data["primaryProfession"].isin(professions_we_want)]
names_data
```

```python
f = open("names.csv", 'w', encoding='utf-8')
names_data.to_csv(f, encoding='utf-8')
f.close()
```
